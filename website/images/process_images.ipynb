```python
import os
from PIL import Image

paths = ['general_bc', 'general_cogagent', 'general_digirl', 'general_gpt4v', 'webshop_bc', 'webshop_cogagent', 'webshop_digirl', 'webshop_gpt4v']

for path in paths:
    print(path)
    image_folder_path = f"./website/images/{path}"
    # remove all .DS_Store
    if '.DS_Store' in os.listdir(image_folder_path):
        os.remove(os.path.join(image_folder_path, '.DS_Store'))
    # remove all _left images
    for image in os.listdir(image_folder_path):
        if "_left" in image or "_right" in image:
            os.remove(os.path.join(image_folder_path, image))
    # check how many images are in the folder
    images = os.listdir(image_folder_path)
    # only count .jpg files
    images = [image for image in images if ".jpg" in image]
    for i, image in enumerate(images):
        print(image)
        image_path = os.path.join(image_folder_path, str(i) + ".jpg")
        # mkdir left image path
        new_folder = os.path.join("./website/images_left/", path)
        if not os.path.exists(new_folder):
            os.makedirs(new_folder)
        # mkdir right image path
        new_folder = os.path.join("./website/images_right/", path)
        if not os.path.exists(new_folder):
            os.makedirs(new_folder)
        # Define the regions for cropping
        left_image_region = (0, 0, 1080, 2280)
        right_image_region = (1080*2, 0, 1080*3, 2280)

        # Open and crop the left image
        image = Image.open(image_path)
        left_image = image.crop(left_image_region)
        # save as 1_left.jpg
        left_image.save(f"./website/images_left/{path}/{str(i)}_left.jpg")

        # Open and crop the right image
        right_image = image.crop(right_image_region)
        # save as 1_right.jpg
        right_image.save(f"./website/images_right/{path}/{str(i)}_right.jpg")
```
